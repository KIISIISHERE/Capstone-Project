```python
# Import libraries
import matplotlib.pyplot as plt
# Data from the table
block_types = ['Plastic Block', 'Wooden Block', 'Metal Block']
mass_g = [100, 200, 500]           # in grams
force_n = [1.0, 2.0, 5.5]          # in Newtons
# Plotting the graph
plt.figure(figsize=(8, 5))
plt.plot(mass_g, force_n, marker='o', linestyle='-', color='blue', label='Frictional Force')
# Annotate each point with the block type
for i, label in enumerate(block_types):
    plt.annotate(label, (mass_g[i], force_n[i] + 0.1))
# Labels and title
plt.title('Frictional Force vs Mass of Block')
plt.xlabel('Mass of Block (g)')
plt.ylabel('Force to Pull Block (N)')
plt.grid(True)
plt.legend()
plt.show()
```
